```python
import pandas as pd
```

```python
file_name = '코로나_naver_news'
# data_df = pd.read_csv('../data/'+file_name+'.csv', encoding='CP949')
data_df = pd.read_csv('../data/'+file_name+'.csv', encoding='UTF-8')
```

```python
data_df.head()
```

```python
description = data_df['description']
description
```

```python
from konlpy.tag import Okt
okt=Okt()
```

```python
description_noun_tk = []
for d in description:
    description_noun_tk.append(okt.nouns(d))
```

```python
description_noun_tk
```

```python
# 제거?
description_noun_tk2 = []

for d in description_noun_tk:
    # 컴프리텐션?
    item=[i for i in d if len(i)>1]
    description_noun_tk2.append(item)
    
    # item=[i for i in d if len(i)>1] 코드랑 같음
    #
    # item=[]
    # for i in d:
    #     if len(i)>1:
    #         item.append(i)

```

```python
description_noun_tk2
```

```python
# !pip install gensim
```

```python
import gensim
import gensim.corpora as corpora
```

```python
dictionary = corpora.Dictionary(description_noun_tk2)
dictionary
```

```python
print(dictionary[0])
print(dictionary[10])
print(dictionary[20])
```

```python
# 단어와 출현빈도 코퍼스 생성
corpus = [dictionary.doc2bow(word) for word in description_noun_tk2]
```

```python
corpus
# (15,2) --> 15번째 단어 2번 출현
```

```python
k = 4   # 토픽의 갯수
```

```python
lda_model = gensim.models.ldamulticore.LdaMulticore(corpus,
                                                    iterations=12,
                                                    num_topics=4,
                                                    id2word=dictionary,
                                                    passes=1,
                                                    workers=10)
                                                
# corpus : 단어(id)별 출현빈도
# id2word : id to word
# num_topics : 생성될 토픽의 개수
# workers : 한번의 트레이닝에 처리될 문서의 개수
# passes :전체 코퍼스 트레이닝 횟수
# iterations : 문서 당 반복 횟수
```

```python
print(lda_model.print_topics(num_topics=k, num_words=10))
```

```python
# 시각화
import pyLDAvis.gensim_models

lda_vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
```
